```python
# We first Import Libraries
import pandas as pd
import plotly.graph_objects as go
```

#### We will Load MonkeyPox extracted in CDC Website

```python

## Mount your Google Drive to Collaboratory
from google.colab import drive
drive.mount('/content/gdrive')

# Loading the CSV file
df = pd.read_csv('gdrive/My Drive/MPX-Data.csv')
df
```

```python
country_list = df["Country"].tolist()
#country_list
country_set = set(country_list)
#country_set
country_list = list(country_set)
#country_list
country_list.sort()
country_list

total_cases = df["Cases"].tolist()
total_cases
```

```python
new_data = pd.DataFrame(list(zip(country_list, total_cases)), columns = ['Country', 'Cases'])
new_data.head()
```

```python
colors = ["#F9F9F5", "#FAFAE6", "#FCFCCB", "#FCFCAE",  "#FCF1AE", "#FCEA7D", "#FCD97D",
          "#FCCE7D", "#FCC07D", "#FEB562", "#F9A648",  "#F98E48", "#FD8739", "#FE7519",
          "#FE5E19", "#FA520A", "#FA2B0A", "#9B1803",  "#861604", "#651104", "#570303",]
```

```python
fig = go.Figure(data=go.Choropleth(
    locationmode = "country names",
    locations = new_data['Country'],
    z = new_data['Cases'],
    text = new_data['Cases'],
    colorscale = colors,
    autocolorscale=False,
    reversescale=False,
    colorbar_title = 'Reported MonkeyPox Cases:: ',
))

fig.update_layout(
    title_text='Reported MonkeyPox Cases :: Hooli Developers (www.hooli.co.tz)',
    geo=dict(
        showcoastlines=True,
    ),
)

#fig.write_html('monkey.html', auto_open=True)
fig.show()
```
